```julia
counter = 0
for index in indices
    
    for coords in index


        ζ = number_states* counter
        # TT -- TT
        TT_energies = [deltaTT12,deltaTT13,deltaTT14,deltaTT23,deltaTT24,deltaTT34]


        TT_TT_coupling_matrix = [
            0    j121 j122 j123 j124 j125;
            j121 0    j126 j127 j128 j129;
            j122 j126 0    j130 j131 j132;
            j123 j127 j130 0    j133 j134;
            j124 j128 j131 j133 0    j135;
            j125 j129 j132 j134 j135 0

        ]   
        ζ 

        for i = Int32(1 + n + n_CT + ζ ) : Int32( n_TT + n + n_CT  + ζ ) 
            for j = Int32(1 + n + n_CT + ζ ) : Int32( n_TT + n + n_CT + ζ ) 
                ip = Int32(i) - (n + n_CT + ζ )
                jp = Int32(j) - (n + n_CT + ζ )
                if ip == jp
                    hamiltonian[i,j] = TT_energies[jp] 
                elseif ip != jp
                    hamiltonian[i,j] = TT_TT_coupling_matrix[ip,jp] 
                
                end
            end
        end        
            


        #### S - S
        S_energies = [delta1,delta2,delta3,delta4]

        S_S_coupling_matrix = [
            0    J1   J2   J3;
            J1   0    J4   J5;
            J2   J4   0    J6;
            J3   J5   J6   0
        ]


        for i = 1 + ζ : n + ζ 
            for j = 1 + ζ :  n + ζ
                 
                ip = i -  ζ 
                jp = j -  ζ 
                if i == j
                    hamiltonian[i,j] = S_energies[jp] 
                

                elseif ip != jp
                    hamiltonian[i,j] =  S_S_coupling_matrix[ip,jp]
                end
            end
        end         


        ## CT - CT


        CT_energies = [deltaCT12,deltaCT13,deltaCT14,deltaCT23,deltaCT24,deltaCT34,deltaCT21,deltaCT31,deltaCT41,deltaCT32,deltaCT42,deltaCT43]

        CT_CT_coupling_matrix = [
            0    J7   J8    J9   J10  J11  J85  J86  J87  J88  J89  J90;
            J7   0    J12  J13  J14  J15  J91  J92  J93  J94  J95  J96;
            J8   J12  0    J16  J17  J18  J97  J98  J99  J100 J101 J102;
            J9   J13  J16  0    J19  J20  J103 J104 J105 J106 J107 J108;
            J10  J14  J17  J19  0    J21  J109 J110 J111 J112 J113 J114;
            J11  J15  J18  J20  J21  0    J115 J116 J117 J118 J119 J120;
            J85  J91  J97  J103 J109 J115 0    J46  J47  J48  J49  J50;
            J86  J92  J98  J104 J110 J116 J46  0    J51  J52  J53  J54;
            J87  J93  J99  J105 J111 J117 J47  J51  0    J55  J56  J57;
            J88  J94  J100 J106 J112 J118 J48  J52  J55  0    J58  J59;
            J89  J95  J101 J107 J113 J119 J49  J53  J56  J58  0    J60;
            J90  J96  J102 J108 J114 J120 J50  J54  J57  J59  J60  0
        ]


        n_CT = Int32(n_CT)
        n = Int32(n)
        for i = (1 + n + ζ ):(n_CT + n + ζ )
            for j =(1 + n + ζ ):(n_CT + n + ζ  )
                ip = i - n - ζ
                jp = j - n - ζ
                if ip == jp
                    hamiltonian[i,j] = CT_energies[jp]    
                else
                    hamiltonian[i,j] = CT_CT_coupling_matrix[ip,jp]        
                end

                
            end

        end          




        # TT - S

        S_TT_coupling_matrix = [
            j136 j142 j148 j154;
            j137 j143 j149 j155;
            j138 j144 j150 j156;
            j139 j145 j151 j157;
            j140 j146 j152 j158;
            j141 j147 j153 j159
        ]


        n = Int32(n)
        n_TT = Int32(n_TT)


        for i = (1 + n + n_CT + ζ ):(n_TT+ n +n_CT + ζ )
            for j = 1+ ζ : n+ ζ 
                ip = i - (n + n_CT + ζ )
                jp = j - ζ
                if ip != jp
                    hamiltonian[i,j] = S_TT_coupling_matrix[ip,jp]
                end
            end
        end


        # S - TT

        TT_S_coupling_matrix = transpose(S_TT_coupling_matrix)

        for i = 1 + ζ : n + ζ 
            for j = (1+n+n_CT+ ζ ):(n_TT+n+n_CT+ ζ )
                ip = i - ζ
                jp = j - (n + n_CT + ζ )
                if ip != jp
                    hamiltonian[i,j] =  TT_S_coupling_matrix[ip,jp]
                end
            end
        end



        ############
        #S - CT
        S_CT_coupling_matrix = [
            J22  J23  J24  J25  J26  J27  J61  J62  J63  J64  J65  J66;
            J28  J29  J30  J31  J32  J33  J67  J68  J69  J70  J71  J72;
            J34  J35  J36  J37  J38  J39  J73  J74  J75  J76  J77  J78;
            J40  J41  J42  J43  J44  J45  J79  J80  J81  J82  J83  J84
        ]


        for i = 1+ ζ :n + ζ 
            for j = (1 + n + ζ ):(n_CT + n + ζ )
                ip = i - ζ
                jp = j -n -ζ
                if ip != jp
                    hamiltonian[i,j] = S_CT_coupling_matrix[ip,jp]
                    
                end
            end
        end

        # CT - s
        CT_S_coupling_matrix = transpose(S_CT_coupling_matrix)

        for i = (1 + n + ζ ):(n_CT + n+ ζ )
            for j = 1+ ζ : n + ζ 
                ip = i - n - ζ
                jp = j - ζ
                if ip != jp
                    hamiltonian[i,j] = CT_S_coupling_matrix[ip,jp]
                    
                end
            end
        end







        # CT - TT

        TT_CT_coupling_matrix = [
            j160 j166 j172 j178 j184 j190 j196 j202 j208 j214 j220 j226;
            j161 j167 j173 j179 j185 j191 j197 j203 j209 j215 j221 j227;
            j162 j168 j174 j180 j186 j192 j198 j204 j210 j216 j222 j228;
            j163 j169 j175 j181 j187 j193 j199 j205 j211 j217 j223 j229;
            j164 j170 j176 j182 j188 j194 j200 j206 j212 j218 j224 j230;
            j165 j171 j177 j183 j189 j195 j201 j207 j213 j219 j225 j231
        ]

        CT_TT_coupling_matrix = transpose(TT_CT_coupling_matrix)
        n = Int32(n)
        n_CT = Int32(n_CT)
        n_TT = Int32(n_TT)

        for i = (1 + n + n_CT + ζ ):(n_TT + n_CT + n + ζ )
            for j = ( 1+ n + ζ ):( n_CT + n + ζ  )
                ip = i - ( n + n_CT + ζ ) 
                jp = j - (n+ ζ )
                if ip != jp
                    hamiltonian[i,j] = TT_CT_coupling_matrix[ip,jp]
                end
            end
        end

        # CT - TT

        for i = (1 + n + ζ ):( n_CT + n + ζ )
            for j = (1 + n + n_CT + ζ ):(n_TT + n_CT + n + ζ )
                ip = i - ( n + ζ  ) 
                jp = j - (n + n_CT + ζ )
                if ip != jp
                    hamiltonian[i,j] = CT_TT_coupling_matrix[ip,jp]
                    
                end
            end
        end
    

    end

    counter += 1
end





```

```julia
#findig characters
# 1 - n = s
# n +1 - n_CT + n = ct

# 1 + n + n_CT - n+ n_CT + n_TT = tt
eig_val = eigen(hamiltonian).values
eig_vec = eigen(hamiltonian).vectors 

 idx = sortperm(eig_val)  
 eig_val = eig_val[idx]
 eig_vec = eig_vec[:,idx]

ehamiltonian = eig_vec

CT_char = []
S_char = []
TT_char = []
eigenvector_charlist = []
n = Int32(n)
n_TT = Int32(n_TT)
n_CT = Int32(n_CT)



for i = 1 : total_states

    S_char_i = dot(ehamiltonian[:,i][1:n],ehamiltonian[:,i][1:n])
    CT_char_i =  dot(ehamiltonian[:,i][(1+n):(n+n_CT)],ehamiltonian[:,i][(1+n):(n+n_CT)])
    TT_char_i =  dot(ehamiltonian[:,i][(1 + n + n_CT):(n + n_CT + n_TT)],ehamiltonian[:,i][(1 + n + n_CT):(n + n_CT + n_TT)])

    eigenvector_charlist_i = [S_char_i,CT_char_i,TT_char_i]
    push!(eigenvector_charlist,(eigenvector_charlist_i))
    push!(CT_char,CT_char_i)
    push!(S_char,S_char_i)
    push!(TT_char,TT_char_i)
end
```

```julia
# the plots of oscillator energy strengths


μ = []
ϕ = pi/4 # phase
β = 0



for j in 1:total_states
        κ = [sin(β + oddeven(j)*deg2rad(29.9))*cos(ϕ), sin(β + oddeven(j)*deg2rad(29.9) )*sin(ϕ),cos(β + oddeven(j)*deg2rad(29.9))]
        ζ = κ./norm(κ) # normalising the stuff
        if j > n
            ζ = zeros(3) ;
        end    

        push!(μ,ζ)
end


TDM_list = []
osc_list = []
for i = 1 : total_states

    TDM_i = ehamiltonian[:,i]' * μ
    push!(TDM_list,TDM_i)
    osc_i = dot(TDM_i,TDM_i)
    push!(osc_list,osc_i)
    
end
osc_list = osc_list/maximum(osc_list)
;


```
